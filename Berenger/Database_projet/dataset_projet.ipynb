```python
import numpy as np
import pandas as pd
```

```python
basics_df = pd.read_csv("https://datasets.imdbws.com/title.basics.tsv.gz", sep="\t", low_memory=False)
basics_df.head(1)
```

```python
ratings_df = pd.read_csv("https://datasets.imdbws.com/title.ratings.tsv.gz", sep="\t", low_memory=False)
ratings_df.head(1)
```

```python
akas_df = pd.read_csv("https://datasets.imdbws.com/title.akas.tsv.gz", sep="\t", dtype={"titleId": "string", "ordering": str, "title ": "string", "region ": "string", 
"language ": "string", "types ": str, "attributes ": str, "isOriginalTitle": str})
akas_df.head()
```

```python
akas_df = akas_df[["titleId", 'region']]
```

```python
region_dataframe = akas_df[akas_df['region'] != '\\N']
region_dataframe
```

```python
region_dataframe.isna().sum()
```

```python
region_dataframe = region_dataframe.dropna()
```

```python
region_dataframe.isna().sum()
```

```python
region_dataframe
```

```python
region_dataframe2 = region_dataframe[region_dataframe['region'].str.contains('FR')]
region_dataframe2

```

```python
region_dataframe2 = region_dataframe2.rename(columns={'titleId': 'tconst'})
region_dataframe2

```

```python
NewDataFramePloum = pd.merge(region_dataframe2, ratings_df, how='left', left_on='tconst', right_on='tconst')
NewDataFramePloum.head()
```

```python
NewDataFrame = pd.merge(basics_df, NewDataFramePloum, how='left', left_on='tconst', right_on='tconst')
NewDataFrame.head()
```

```python
NewDataFrame.loc[NewDataFrame['tconst'] == "tt0119038"]
```

```python
NewDataFrame2 = NewDataFrame.loc[NewDataFrame['titleType'] == "movie"]
NewDataFrame3 = NewDataFrame2[NewDataFrame2['startYear'] != '\\N']
```

```python
NewDataFrame4 = NewDataFrame3.astype({"startYear": int})
NewDataFrame5 = NewDataFrame4[NewDataFrame4['startYear'] > 1960]
NewDataFrame5.head()
```

```python
NewDataFrame4 = NewDataFrame3.astype({"startYear": int})
NewDataFrame5 = NewDataFrame4.loc[(NewDataFrame4['startYear'] >= 1960) & (NewDataFrame5['startYear'] <= 2022)]
```

```python
NewDataFrame5.sort_values(by=['startYear'], ascending=True)
```

```python
df_2022_2021 = NewDataFrame5.loc[NewDataFrame5['startYear'] > 2020]
df_2022_2021_2 =  df_2022_2021.loc[df_2022_2021['region'] == 'FR']
df_2022_2021_2
```

```python
NewDataFrame5.loc[NewDataFrame5['tconst'] == "tt0119038"]
```

```python

NewDataFrame6 = NewDataFrame5.loc[(NewDataFrame5['startYear'] <= 2020) & (NewDataFrame5['numVotes'] >= 25000)]
NewDataFrame6
```

```python
NewDataFrame6.loc[NewDataFrame6['tconst'] == "tt0119038"]
```

```python
frames = [df_2022_2021_2, NewDataFrame6]
frames2 = pd.concat(frames)
frames2
```

```python
frames3 = frames2.drop_duplicates(subset=['tconst'], keep=False)
frames3
```

```python
frames3 = frames3[["primaryTitle", "startYear", "genres", "averageRating", "numVotes"]]
frames3
```

```python
new = frames3["genres"].str.split(",", n = 1, expand = True)
new2 = new[1].str.split(",", n = 1, expand = True)
frames3["genre1"] = new[0]
frames3["genre2"] = new2[0]
frames3["genre3"] = new2[1]
frames4 = frames3[["primaryTitle", "startYear", "genre1", "genre2", "genre3", "averageRating", "numVotes"]]
frames4
```

```python
frames5 = pd.concat([frames4['genre1'].str.get_dummies(), frames4['genre2'].str.get_dummies(), frames4['genre3'].str.get_dummies()], axis = 1)
frames5
```

```python
frames6 = pd.concat([frames4, frames5.groupby(lambda x:x, axis=1).sum()], axis = 1)
frames6
```
