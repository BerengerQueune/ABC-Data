```python
import numpy as np
import pandas as pd
```

```python
name_DF = pd.read_csv('D:/Data analyst/Wild code school/Projets/Projet 2 - Pandas/BD/name.basics.tsv', sep = '\t')
```

```python
name_DF.head()
```

```python
#Filtrer les acteurs
nameActors = name_DF[(name_DF['primaryProfession'].str.contains('actor', na = False)) 
                  | (name_DF['primaryProfession'].str.contains('actress', na = False))]
```

```python
#Filtrer les vivants et ceux dont on connai la date de naissance
vivActeurs = nameActors[(nameActors['deathYear'] == r'\N') & (nameActors['birthYear'] != r'\N')]
```

```python
vivActeurs.head(10)
```

```python
#ajouter une colonne "age" en prenant l'écart entre la date de naissance et la dte actuelle
vivActeurs['birthYear'] = pd.to_numeric(vivActeurs['birthYear'])
vivActeurs = vivActeurs.assign(age = 2021 - vivActeurs['birthYear'])
```

```python
#ajoute une colonne sur le sexe
vivActeurs = vivActeurs.assign(sexe = 'M')
```

```python
#ajoute une colonne sur le sexe
def sexe(pp):
    if 'actor' in pp:
        return 'M'
    else:
        return 'F'
    
vivActeurs['sexe'] = vivActeurs['primaryProfession'].apply(sexe)
vivActeurs.head(10)
```

```python
moy = round(vivActeurs['age'].mean(),1)

dfFemme = vivActeurs[vivActeurs['sexe'] == 'F']
nbFemme = dfFemme['sexe'].count()
moyFemme = round(dfFemme['age'].mean(),1)

dfHomme = vivActeurs[vivActeurs['sexe'] == 'M']
nbHomme = dfHomme['sexe'].count()
moyHomme = round(dfHomme['age'].mean(),1)

print(f"La moyenne d'age des actrices et acteurs vivants est de {moy} ans.")
print(f"Il y a {nbFemme} actrices vivantes avec une moyenne d'age de {moyFemme} ans et {nbHomme} acteurs vivants avec une moyenne d'age de {moyHomme} ans.")
```
