```python
import pandas as pd
import numpy as np
```

```python
title = pd.read_csv('D:/Data analyst/Wild code school/Projets/Projet 2 - Pandas/BD/Christophe_DataFrame.csv', sep = ',')
```

```python
film1 = title[title.titleType.isin (['movie', 'tvMovie'])]
```

```python
film2 = film1[film1["genres"].apply(lambda x:x not in ['short'])]
```

```python
lang = pd.read_csv('D:/Data analyst/Wild code school/Projets/Projet 2 - Pandas/BD/title.akas.tsv', sep = '\t')
```

```python
lang1 = lang[lang['region']=='FR']
```

```python
merge_L_Y = lang1.merge(film2, how = "inner", left_on = "titleId", right_on = "tconst")
```

```python
merge_L_Y['language'].value_counts() 
```

```python
merge_L_Y.to_csv (r'D:\Data analyst\Wild code school\Projets\Projet 2 - Pandas\BaseCL1.csv', index = False, header=True)
```
