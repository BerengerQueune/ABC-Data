```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
basic_DF = pd.read_csv('D:/Data analyst/Wild code school/Projets/Projet 2 - Pandas/BD/Christophe_DataFrame.csv', sep = ',')
```

```python
rating_DF = pd.read_csv('D:/Data analyst/Wild code school/Projets/Projet 2 - Pandas/BD/title.ratings.tsv', sep = '\t')
```

```python
#Je regarde la volumétrie
rating_DF[(rating_DF['averageRating'] > 9) & (rating_DF['numVotes']>=100000)].count()
```

```python
#Je sélectionne selon les critères ci dessus
rating2 = rating_DF[(rating_DF['averageRating'] > 9) & (rating_DF['numVotes']>=100000)] 
```

```python
rating2
```

```python
#Je merge avec la table des films pour récupérer les noms
rating3 = rating2.merge(basic_DF, how = 'left', left_on = 'tconst', right_on = 'tconst')
```

```python
rating3 = rating3.sort_values(['averageRating'], ascending = False)
rating3
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=rating3['primaryTitle'], y=rating3['averageRating'])
plt.xticks(rotation= 70)
plt.xlabel('primaryTitle')
plt.ylabel('averageRating')
plt.title('FIlms les mieux notés avec plus de 100 000 votes')
```
